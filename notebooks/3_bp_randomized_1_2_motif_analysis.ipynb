```python
bps_wobble = []
for index in score.index:
    bp = ''
    for pos in range(0, 3):
        left = index[pos]
        right = index[::-1][pos]
        
        if sorted([left, right]) in [sorted(['A', 'U']), sorted(['G', 'C'])]: bp += 'p'
        elif sorted([left, right]) == sorted(['G', 'U']): bp += 'w'
        else: bp += 'm'
    bps_wobble.append(bp)
            
bps_no_wobble = []
for bp in bps_wobble:
    if bp == 'mwm': bp = 'mmm'
    else: bp = bp.replace('w', 'p')
    bps_no_wobble.append(bp[::-1])

score.loc[:, 'bp'] = bps_no_wobble
score
```

```python
import sys
sys.path.append('/casa/haedong/tools')
import plotutils
from matplotlib.ticker import AutoMinorLocator
```

```python
if RANDOMIZED == 2:
    bp2color = {
        'ppm': '#a22100',
        'ppp': '#f15a34',
        'pmp': '#efc150',
        'pmm': '#9cc24b',
        'mpp': '#64c496',
        'mpm': '#60c1f2',
        'mmp': '#3889c0',
        'mmm': '#5b5c7b'
    }
    
elif RANDOMIZED == 1:
    bp2color = {
        'mpp': '#a22100',
        'ppp': '#f15a34',
        'ppm': '#efc150',
        'mpm': '#9cc24b',
        'pmp': '#64c496',
        'pmm': '#60c1f2',
        'mmp': '#3889c0',
        'mmm': '#5b5c7b'
    }
    
else: raise ValueError()
```

```python
for condition in score.columns[score.columns.str.startswith('cleavage_score')]:
    fig, ax = plt.subplots(1, 1, figsize=(3, 2))

    bp2average = {}
    for key, grp in score.groupby('bp'):
        bp2average[key] = grp[condition].mean()
        
    n = 100
    for bp in sorted(bp2average, key=bp2average.get, reverse=True):
        sorted_data = np.sort(np.array(score.loc[score['bp'] == bp][condition]))
        ax.step(np.concatenate([sorted_data, sorted_data[[-1]]]), (np.arange(sorted_data.size + 1) / sorted_data.size), color=bp2color[bp], label=bp, zorder=n)
        n -= 1
        
    ax.set_xlabel('Cleavage score')
    ax.set_ylabel('Cumulative fraction')
    ax.set_title(condition)

    ax.xaxis.grid(False)
    plotutils.adjust_numbers_style(ax, xgrid=True, ygrid=False)
    ax.xaxis.set_minor_locator(AutoMinorLocator(4))
    plt.setp(ax.xaxis.get_minorticklines(), visible=False)
    plt.grid(True, which='minor', color='#d0d0d0', linestyle='-', linewidth=0.5)

    legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
    for text in legend.get_texts():
        text.set_color('black')
        
    #plt.savefig('{}/cumulative.{}.pdf'.format(FIG_DIR, condition), format='pdf', bbox_inches='tight')
    plt.show()
```

```python
import seaborn as sns
sns.set(style='whitegrid', font='Helvetica')
```

```python
if RANDOMIZED == 2: bp_of_interest = ('ppm', 'ppp')
elif RANDOMIZED == 1: bp_of_interest = ('mpp', 'ppp')
else: raise ValueError()

for condition in score.columns.drop(['bp', '1_position']):
    dataframes = []
    for base_3p in ['G', 'C', 'A', 'U']:
        lines = []
        for base_5p in ['G', 'C', 'A', 'U']:
            lines.append([base_5p, score.loc[(score['bp'].isin(bp_of_interest)) & (score['1_position'] == '{}/{}'.format(base_5p, base_3p))][condition].mean()])
        df = pd.DataFrame(lines)
        df.columns = ['base_5p', base_3p]
        df = df.set_index('base_5p')
        dataframes.append(df)
    df = pd.concat(dataframes, axis=1)
    
    fig, ax = plt.subplots(1, 1, figsize=(2.5, 2))

    sns.heatmap(df, square=False, linewidths=0.1, cmap='viridis', annot=False,
                xticklabels=True, yticklabels=True, cbar=True, ax=ax)
    
    plt.savefig('{}/1_position.{}.randomized_{}.{}_cleavage.linear_scale.pdf'.format(FIG_DIR, RNA, RANDOMIZED, condition), format='pdf', bbox_inches='tight')
    plt.show()
```
